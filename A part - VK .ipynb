Есть 100 значений некоторой абстрактной переменной, по которым была обучена модель. Затем была посчитана среднеквадратическая ошибка (MSE). Для 80 случаев разница истинного и предсказанного значения составляет 0,5, а для остальных 20 случаев равняется -0,3. Вы знаете, что модель была обучена неверно. Если добавить к ответам модели некоторую константу C, то результат улучшится. Существует ли такая константа? Если да, найдите её.

**Ответ:** MSE не может быть отрицательной. Потому-что MSE измеряет среднеквадратичную ошибку прогнозов, по формуле:  (сумма квадратов ошибок объектов / количество объектов) и возведение значений в квадрат как раз избавляет от разницы между переоценкой и недооценкой модели (отрицательных значений). 

Зы. Нужно проверить как у нас считается MSE, в sklearn есть встроенная для этого функция или можно написать функцию самому: 


```python
def mse(answers, predictions):
    total = 0
    for i in range(len(answers)):
        total+=(predictions[i]-answers[i])**2
    result = total / len(answers)
    return result
```
